## ROLLING MEAN WITH ROLLING STD - BOLLINGER'S BAND

```python
dates = pd.date_range('2015-01-01','2016-12-31')
df = get_data(symbols,dates)
moving_average = pd.rolling_mean(df['GOOG'], 20, min_periods=None, freq=None, center=False, how=None)
rolling_std    = pd.rolling_std(df['GOOG'], 20, min_periods=None, freq=None, center=False, how=None)

upp_std       =  moving_average + 2*rolling_std
down_std      =  moving_average - 2*rolling_std

upp_std.plot(ls='--')
down_std.plot(ls='--')
moving_average.plot()
df['GOOG'].plot()
plt.show()
```

## DAILY RETURNS

```python


def daily_returns(df):
    daily = df.copy()
    daily[1:] = (daily[1:]/daily[:-1].values)-1
    daily.iloc[0,:]=0
    return(daily)

dates = pd.date_range('2015-01-01','2015-01-31')
df = get_data(symbols,dates)
plot_data(daily_returns(df))
```

## Cummulative Returns

```python

def Cummilative_returns(df):
    daily = df.copy()
    daily[1:] = (daily/daily.iloc[1])-1
    daily.iloc[0,:]=0
    return(daily)
    
dates   = pd.date_range('2015-01-01','2018-01-31')
symbols = ['IBM']
df = get_data(symbols,dates)
plot_data(Cummilative_returns(df))
```
